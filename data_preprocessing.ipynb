```python
import pandas as pd
import numpy as np
```

```python
feed_data = pd.read_csv('./feed_data_prep.csv')
user_data = pd.read_csv('./user_data_prep.csv')
post_data = pd.read_csv('./post_data_prep.csv')
```

```python
data = pd.merge(feed_data, user_data, on='user_id')
data = pd.merge(data, post_data, on='post_id')
```

```python
data['city'] = data.groupby('city')['target'].transform('mean')
data['country'] = data.groupby('country')['target'].transform('mean')
data['topic'] = data.groupby('topic')['target'].transform('mean')
```
